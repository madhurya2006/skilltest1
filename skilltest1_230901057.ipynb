```python
#1a
import numpy as np
import pandas as pd

# Sample exam scores for 20 students
scores = [78, 45, 88, 92, 67, 59, 73, 84, 90, 33,
          49, 60, 77, 85, 52, 41, 69, 81, 38, 95]

# Create a DataFrame
df = pd.DataFrame({'Score': scores})

# 1. Overall class performance
average_score = np.mean(df['Score'])
highest_score = np.max(df['Score'])
lowest_score = np.min(df['Score'])

print(" Overall Class Performance ")
print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score: {lowest_score}")
print()

# 2. Pass/Fail Classification
pass_mark = 50
df['Result'] = np.where(df['Score'] >= pass_mark, 'Pass', 'Fail')

num_passed = (df['Result'] == 'Pass').sum()
num_failed = (df['Result'] == 'Fail').sum()

print("Pass/Fail Classification")
print(f"Number of Students Passed: {num_passed}")
print(f"Number of Students Failed: {num_failed}")
print()

# 3. Grading on a Curve (Using Fixed Thresholds)
# A: 90–100, B: 80–89, C: 70–79, D: 60–69, F: <60

def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Score'].apply(assign_grade)

print("Grades Assigned")
for idx, row in df.iterrows():
    print(f"Student {idx+1}: Score = {row['Score']}, Grade = {row['Grade']}, Result = {row['Result']}")
```

```python
#2b
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load data from CSV file
df = pd.read_csv('/content/grades1 - Sheet1.csv')  # CSV file must have Name, Math, Science, English columns

# 2. Compute total and average scores
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# 3. Subject-wise statistics
subjects = ['Math', 'Science', 'English']
print("Subject-wise Statistics")
for subject in subjects:
    avg = df[subject].mean()
    high = df[subject].max()
    low = df[subject].min()
    print(f"{subject} - Avg: {avg:.2f}, Highest: {high}, Lowest: {low}")

# 4. Top 5 students by total score
print("\n Top 5 Students by Total Score ")
top_students = df.sort_values(by='Total', ascending=False).head(5)
for _, row in top_students.iterrows():
    print(f"{row['Name']}: Total = {row['Total']}, Average = {row['Average']:.2f}")

# 5. Plot histograms for each subject
plt.figure(figsize=(12, 4))
for i, subject in enumerate(subjects):
    plt.subplot(1, 3, i + 1)
    plt.hist(df[subject], bins=8, color='lightblue', edgecolor='black')
    plt.title(f'{subject} Grades')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#3a
import numpy as np

# Functions to calculate impedance of each component
def get_resistor_impedance(R):
    return complex(R)

def get_inductor_impedance(L, f):
    return complex(0, 2 * np.pi * f * L)

def get_capacitor_impedance(C, f):
    return complex(0, -1 / (2 * np.pi * f * C))

# Functions to calculate total impedance
def series_impedance(impedances):
    return sum(impedances)

def parallel_impedance(impedances):
    return 1 / sum(1 / z for z in impedances)

# ==== Example values (you can change these) ====
frequency = 50  # in Hz

# Component values
R = 100        # ohms
L = 0.01       # henrys (10 mH)
C = 100e-6     # farads (100 µF)

# Get individual impedances
Z_R = get_resistor_impedance(R)
Z_L = get_inductor_impedance(L, frequency)
Z_C = get_capacitor_impedance(C, frequency)

# Put all components in a list
impedance_list = [Z_R, Z_L, Z_C]

# Choose configuration: "series" or "parallel"
configuration = "series"  # or "parallel"

# Calculate total impedance
if configuration == "series":
    Z_total = series_impedance(impedance_list)
else:
    Z_total = parallel_impedance(impedance_list)

# Print result
print(f"\nTotal Impedance ({configuration.title()}): {Z_total:.2f} Ω")
print(f"  Magnitude: {abs(Z_total):.2f} Ω")
print(f"  Phase angle: {np.angle(Z_total, deg=True):.2f}°")
```

```python
#4a
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sampling_rate = 50  # Hz
duration = 1  # in seconds
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
def generate_wave(freq, amp, noise_level):
    signal = amp * np.sin(2 * np.pi * freq * t)
    noise = np.random.normal(0, noise_level, t.shape)
    return signal + noise
data={
    'Patient_ID':[f"{i}" for i in range(1,51)],
    'Time(s)':t,
    'Delta': generate_wave(2, 10, 10),
    'Theta': generate_wave(6, 6, 8),
    'Alpha': generate_wave(10, 4, 5),
    'Beta':  generate_wave(20, 5, 3)
}
df = pd.DataFrame(data)
df.to_csv(r"C:\Users\DELL\Downloads\EEG-Sheet1.csv",index=False)
print(df)
print("\nStatistics (Mean, Variance, Max):")
print('Mean:\n',df[['Delta', 'Theta', 'Alpha', 'Beta']].mean())
print('Variance:\n',df[['Delta', 'Theta', 'Alpha', 'Beta']].var())
print('Max:\n',df[['Delta', 'Theta', 'Alpha', 'Beta']].max())
plt.bar(df['Patient_ID'],df['Delta'])
plt.xlabel('Patient_ID')
plt.ylabel('Delta')
plt.show()
plt.bar(df['Patient_ID'],df['Theta'])
plt.xlabel('Patient_ID')
plt.ylabel('Theta')
plt.show()
plt.bar(df['Patient_ID'],df['Alpha'])
plt.xlabel('Patient_ID')
plt.ylabel('Alpha')
plt.show()
plt.bar(df['Patient_ID'],df['Beta'])
plt.xlabel('Patient_ID')
plt.ylabel('Beta')
plt.show()
```
